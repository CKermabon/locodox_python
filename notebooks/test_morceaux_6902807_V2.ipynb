```python
import pandas as pd

def piecewise_model(var1, delta_T, *params):
    """ Modèle linéaire par morceaux avec continuité assurée """
    G = params[0]  # Coefficient global
    D = params[1]  # Paramètre de variation
    breakpoints = params[2:]  # Liste des breakpoints
    
    Y = G * (1 + D * delta_T / (365 * 100)) * var1
    
    # Ajout des segments de régression
    for i, bp in enumerate(breakpoints):
        Y += params[i + 2] * np.maximum(0, var1 - bp)

    return Y

def fit_piecewise(var1, var2, delta_T,breakpoints):
    """ Ajuste G et D avec régression segmentée """
    
    
    # Initialisation des paramètres (G=1, D=0, pentes=0)
    p0 = [1, 0] + [0] * len(breakpoints)
    
    # Ajustement via curve_fit
    popt, _ = curve_fit(lambda v1, *p: piecewise_model(v1, delta_T, *p), var1, var2, p0=p0)
    
    return popt, breakpoints

mask = (np.isfinite(PPOX1)) & (np.isfinite(NCEP_PPOX))
params, breakpoints = fit_piecewise(PPOX1[mask], NCEP_PPOX[mask], delta_T_NCEP[mask],[150])
```

```python
def apply_corrected_model(var1, delta_T, G, D, breakpoints, p_opts):
    """Applique la correction en respectant l'équation originale."""
    var2_corr = np.zeros_like(var1)
    
    # Avant le premier breakpoint
    mask = var1 < breakpoints[0]
    print(len(delta_T[mask]))
    var2_corr[mask] = G * (1 + D * delta_T[mask] / (365 * 100)) * var1[mask]
    
    # Appliquer les corrections sur les segments suivants
    D_corr = D  # Initialisation du D corrigé
    for i, bp in enumerate(breakpoints):
        mask = var1 >= bp
        D_corr += p_opts[i]  # Mise à jour de D après chaque breakpoint
        var2_corr[mask] = G * (1 + D_corr * delta_T[mask] / (365 * 100)) * var1[mask]

    return var2_corr
    
var2_corrected = apply_corrected_model(PPOX1, delta_T_NCEP, params[0], params[1], [150], [params[2]])
```

```python
import argopy
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
fic_argo_Sprof = '/Users/chemon/ARGO_NEW/NEW_LOCODOX/data_test/6902802/6902802_Sprof.nc'
ds = xr.open_dataset(fic_argo_Sprof,engine='argo')
dss = ds[['TEMP','PSAL','DOXY','PRES']]
dss=dss.assign_coords({'N_PROF':range(len(dss.N_PROF)), 'N_LEVELS':range(len(dss.N_LEVELS))})
dssi = dss.argo.interp_std_levels(np.arange(0,1000,100))
```

```python
i=0
_=plt.plot(dss.isel(N_PROF=i)['DOXY'],dss.isel(N_PROF=i)['PRES'],'.-')
_=plt.plot(dssi.isel(N_PROF=i)['DOXY'],dssi.isel(N_PROF=i)['PRES_INTERPOLATED'],'ro')
plt.gca().invert_yaxis()
```
